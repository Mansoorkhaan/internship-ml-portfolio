```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\Administrator\Documents\Mansoor's Documents\Datasets\household_power_consumption.csv", sep=",", low_memory=False)
df.head()
```

```python
# Try flexible datetime parsing
if "Date" in df.columns and "Time" in df.columns:
    df["datetime"] = pd.to_datetime(df["Date"] + " " + df["Time"], errors="coerce", dayfirst=True)
elif "datetime" in df.columns:
    df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")
else:
    # guess first col
    df["datetime"] = pd.to_datetime(df.iloc[:,0], errors="coerce")

df = df.dropna(subset=["datetime"]).set_index("datetime").sort_index()
# Main target (Global_active_power is typical)
target_col = None
for cand in ["Global_active_power", "global_active_power", "power", "consumption"]:
    if cand in df.columns:
        target_col = cand
        break
if target_col is None:
    raise ValueError("Target column not found. Please adjust manually.")

df[target_col] = pd.to_numeric(df[target_col], errors="coerce")
df = df.dropna(subset=[target_col])

print("Data shape:", df.shape)
print("Target column:", target_col)
```

```python
#RESAMPLE (hourly mean)
y = df[target_col].resample("H").mean().fillna(method="ffill")
print("Resampled shape:", y.shape)

plt.figure(figsize=(10,4))
y.plot(title="Hourly Energy Consumption")
plt.tight_layout(); plt.show()
```

```python
#  TRAIN/TEST SPLIT
split_idx = int(len(y)*0.8)
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
```

```python
#  ARIMA MODEL
try:
    arima = ARIMA(y_train, order=(2,1,2))
    arima_fit = arima.fit()
    arima_pred = arima_fit.forecast(steps=len(y_test))
except Exception as e:
    print("ARIMA failed:", e)
    arima_pred = pd.Series(np.repeat(y_train.mean(), len(y_test)), index=y_test.index)
```

```python
#  PROPHET MODEL
prophet_df = y_train.reset_index()
prophet_df.columns = ["ds", "y"]

prophet = Prophet(daily_seasonality=True, yearly_seasonality=True)
prophet.fit(prophet_df)

future = pd.DataFrame({"ds": y_test.index})
forecast = prophet.predict(future)
prophet_pred = forecast["yhat"].values
prophet_pred = pd.Series(prophet_pred, index=y_test.index)
```

```python
#  XGBOOST MODEL
# Feature engineering
def make_features(series):
    df_feat = pd.DataFrame({"y": series})
    df_feat["hour"] = df_feat.index.hour
    df_feat["dayofweek"] = df_feat.index.dayofweek
    df_feat["month"] = df_feat.index.month
    return df_feat

train_feat = make_features(y_train)
test_feat  = make_features(y_test)

X_train, y_train_xgb = train_feat.drop("y", axis=1), train_feat["y"]
X_test, y_test_xgb   = test_feat.drop("y", axis=1), test_feat["y"]

xgb = XGBRegressor(n_estimators=300, max_depth=5, learning_rate=0.05, random_state=42)
xgb.fit(X_train, y_train_xgb)
xgb_pred = pd.Series(xgb.predict(X_test), index=y_test.index)
```

```python
#  EVALUATION
def eval_model(name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{name} -> MAE: {mae:.4f}, RMSE: {rmse:.4f}")
    return mae, rmse

print("\n=== MODEL PERFORMANCE ===")
mae_arima, rmse_arima = eval_model("ARIMA", y_test, arima_pred)
mae_prophet, rmse_prophet = eval_model("Prophet", y_test, prophet_pred)
mae_xgb, rmse_xgb = eval_model("XGBoost", y_test, xgb_pred)
```

```python
#  VISUALIZATION
plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test.values, label="Actual", color="black", linewidth=2)
plt.plot(y_test.index, arima_pred, label="ARIMA Forecast", alpha=0.8)
plt.plot(y_test.index, prophet_pred, label="Prophet Forecast", alpha=0.8)
plt.plot(y_test.index, xgb_pred, label="XGBoost Forecast", alpha=0.8)
plt.title("Energy Consumption Forecast vs Actual")
plt.xlabel("Time"); plt.ylabel("Consumption")
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
```

```python
#  SAVE RESULTS (optional)
out = pd.DataFrame({
    "actual": y_test,
    "arima": arima_pred,
    "prophet": prophet_pred,
    "xgboost": xgb_pred
})
out.to_csv("energy_forecasts.csv")
print("\nSaved: energy_forecasts.csv")
```
