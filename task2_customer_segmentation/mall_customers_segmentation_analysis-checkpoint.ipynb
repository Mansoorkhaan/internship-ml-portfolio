# Task 2 â€” Customer Segmentation (K-Means) + EDA + PCA/t-SNE + Strategies
# Dataset: Mall Customers (CSV)

```python
import re
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv( r"C:\Users\Administrator\Documents\Mansoor's Documents\Datasets\Mall_Customers.csv")
print("Shape:", df.shape)
df.head()
```

```python
# Normalize column names (lowercase, strip, collapse spaces, remove punctuation) ....
def norm(s):
    s = s.strip().lower()
    s = re.sub(r"[^\w\s]+", "", s)  # remove punctuation....
    s = re.sub(r"\s+", " ", s)      # collapse spaces....
    return s

df.columns = [norm(c) for c in df.columns]
print("Columns:", list(df.columns))
print(df.head(3))
```

```python
# 2) EDA ....
# ---------
print("\n=== Info ===")
print(df.info())
print("\n=== Missing Values ===")
print(df.isna().sum())

# Basic stats....
print("\n=== Describe (numeric) ===")
print(df.describe())
```

```python
# Quick histograms for common numeric fields if present....
numeric_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
if numeric_cols:
    df[numeric_cols].hist(figsize=(10,6), bins=15)
    plt.suptitle("Feature Distributions", y=1.02)
    plt.tight_layout()
    plt.show()
```

```python
 #PICK FEATURES for "spending habits"
#    Common columns in Mall Customers:
#    - customerid, gender, age, annual income (k$), spending score (1-100)
# --------------------------------------
# Try to auto-detect likely columns:
def find_col(candidates, required=True):
    for cand in candidates:
        base = norm(cand)
        for col in df.columns:
            if base == col:
                return col # try fuzzy contains
    for cand in candidates:
        key = re.sub(r"[^\w]+", "", cand.lower())
        for col in df.columns:
            if key in re.sub(r"[^\w]+", "", col):
                return col
    if required:
        raise ValueError(f"Could not find any of: {candidates}. Columns present: {df.columns}")
    return None

age_col = find_col(["age"], required=False)
income_col = find_col(["annual income (k$)", "annualincome", "annual income"], required=True)
spend_col = find_col(["spending score (1-100)", "spendingscore", "spending score"], required=True)
gender_col = find_col(["gender"], required=False)
```

```python
# Build feature set (Age + Income + Spending are typical)
feature_cols = [c for c in [age_col, income_col, spend_col] if c is not None]
X = df[feature_cols].copy()
```

```python
# Optional: add gender as 0/1 if present
if gender_col is not None:
    X["gender_flag"] = (df[gender_col].astype(str).str.lower().str[0].map({"f":0, "m":1})).fillna(0)

print("\nUsing feature columns:", list(X.columns))
```

```python
#  SCALE + CHOOSE K (elbow + silhouette)
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

ks = range(2, 11)
inertias, silhouettes = [], []
for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = km.fit_predict(Xs)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(Xs, labels))

fig, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].plot(ks, inertias, marker="o")
ax[0].set_title("Elbow (Inertia)")
ax[0].set_xlabel("k"); ax[0].set_ylabel("Inertia"); ax[0].grid(True)

ax[1].plot(ks, silhouettes, marker="o")
ax[1].set_title("Silhouette Score")
ax[1].set_xlabel("k"); ax[1].set_ylabel("Silhouette"); ax[1].grid(True)
plt.tight_layout(); plt.show()
```

```python
# Pick best k by max silhouette (you can override manually)
best_k = ks[int(np.argmax(silhouettes))]
print(f"\nSelected k (by silhouette): {best_k}")

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init="auto")
labels = kmeans.fit_predict(Xs)
df["cluster"] = labels
```

```python
#  CLUSTER PROFILES
profile_cols = [c for c in [age_col, income_col, spend_col, "gender_flag"] if c is not None and c in df.columns]
if not profile_cols:
    print("\n(No profiling columns found in df; check column names.)")
else:
    profiles = df.groupby("cluster")[profile_cols].agg(["mean", "median", "count"])
    print("\n=== Cluster Profiles ===")
    print(profiles)
```

```python
# Simple readable summary
summary = df.groupby("cluster")[feature_cols].mean().round(1)
summary["size"] = df.groupby("cluster").size()
print("\n=== Cluster Means (+ size) ===")
print(summary)
```

```python
#  PCA (2D) VISUALIZATION
pca = PCA(n_components=2, random_state=42)
p2 = pca.fit_transform(Xs)
print("\nPCA explained variance ratio:", np.round(pca.explained_variance_ratio_, 3))

plt.figure(figsize=(7,5))
for c in range(best_k):
    idx = (labels == c)
    plt.scatter(p2[idx,0], p2[idx,1], s=35, alpha=0.8, label=f"Cluster {c}")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.title("PCA (2D) â€” KMeans Clusters")
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
```

```python
#  t-SNE (optional; slower)
try:
    ts = TSNE(n_components=2, random_state=42, init="pca", perplexity=min(30, Xs.shape[0]-1))
    t2 = ts.fit_transform(Xs)
    plt.figure(figsize=(7,5))
    for c in range(best_k):
        idx = (labels == c)
        plt.scatter(t2[idx,0], t2[idx,1], s=35, alpha=0.8, label=f"Cluster {c}")
    plt.title("t-SNE (2D) â€” KMeans Clusters")
    plt.xlabel("t1"); plt.ylabel("t2")
    plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
except Exception as e:
    print("t-SNE skipped (reason):", e)
```

```python
#  MARKETING STRATEGIES (rule-based from cluster means)
means = df.groupby("cluster")[feature_cols].mean()
overall = df[feature_cols].mean()

# Use relative thresholds (quantiles) for robust rules
q = df[feature_cols].quantile([0.25, 0.5, 0.75])
q25, q50, q75 = q.iloc[0], q.iloc[1], q.iloc[2]

def segment_label(row):
    inc = row[income_col]
    sp  = row[spend_col]
    age = row[age_col] if age_col is not None else np.nan

    if inc >= q75[income_col] and sp >= q75[spend_col]:
        return "High-Value Enthusiasts"
    if inc >= q75[income_col] and sp < q50[spend_col]:
        return "Affluent but Cautious"
    if inc < q50[income_col] and sp >= q75[spend_col]:
        return "Value Seekers"
    if (age_col is not None) and (age <= q25[age_col]) and (sp >= q50[spend_col]):
        return "Young Trend Responders"
    if sp < q25[spend_col]:
        return "Low Engagement"
    return "Balanced Regulars"

labels_map = means.apply(segment_label, axis=1)
print("\n=== Segment Labels by Cluster ===")
for c, name in labels_map.items():
    print(f"Cluster {c}: {name}")
```

```python
# Strategy suggestions per segment
def strategies_for(name):
    if name == "High-Value Enthusiasts":
        return [
            "VIP tier with early access & exclusive drops",
            "Premium bundles and loyalty multipliers",
            "Personalized concierge offers via WhatsApp/email"
        ]
    if name == "Affluent but Cautious":
        return [
            "Trust-building content (reviews, guarantees, free returns)",
            "Premium trials or limited-time upgrade credits",
            "High-touch nurture with value explanations"
        ]
    if name == "Value Seekers":
        return [
            "Flash sales, bundles, and value packs",
            "Points-based rewards and referral bonuses",
            "Social-proof heavy creatives & urgency timers"
        ]
    if name == "Young Trend Responders":
        return [
            "Creator/UGC-driven campaigns on social",
            "Limited editions & seasonal styles",
            "Gamified challenges and campus promos"
        ]
    if name == "Low Engagement":
        return [
            "Low-frequency, high-value reminders (coupons)",
            "Win-back journeys with progressive incentives",
            "Preference surveys to tailor future outreach"
        ]
    return [
        "Balanced promotions with light personalization",
        "Occasional bundles and anniversary rewards",
        "Lifecycle emails based on recency/frequency"
    ]

print("\n=== Marketing Strategies by Cluster ===")
for c, name in labels_map.items():
    print(f"\nCluster {c} â€” {name}")
    for s in strategies_for(name):
        print(f" - {s}")
```
